# One Hot Encoding - variables with many categories

```python
j=pd.read_csv('txt.csv',usecols=['Type 1','Type 2'])
```

```python
j.head()
```

```python
len(j['Type 1'].unique())
```

```python
len(j['Type 2'].unique())
```

```python
# let's have a look at how many labels each variable has

for col in j.columns:
    print(col, ': ', len(j[col].unique()), ' labels')
```

```python
j.shape
```

```python
 #let's examine how many columns we will obtain after one hot encoding these variables
    # IN this we drop a column
m=pd.get_dummies(j,drop_first=True)
```

```python
m.shape #  We can observe that from with just 2 categorical features we are getting 34 features with the help of one hot encoding.
```
